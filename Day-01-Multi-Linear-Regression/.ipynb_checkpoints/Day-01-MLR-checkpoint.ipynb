### Auto regression

```python
cars_df
```

```python
cars_df
```

```python
x = cars_df.drop('MPG',axis=1)
```

```python
print(x)
```

```python
y = cars_df[['MPG']]
```

```python
print(y)
```

### Model Building

```python
model = LinearRegression()
```

### Model Training

```python
model.fit(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```

### Model Testing

```python
y_pred = model.predict(x)
```

```python
print(y_pred)
```

```python
mean_absolute_error(y,y_pred)
```

```python
r2_score(y,y_pred)
```

### homoscadasticity

```python
error = y-y_pred
```

```python
plt.scatter(y,error)
```

### Zero residual mean

```python
plt.scatter(y,y_pred)
```

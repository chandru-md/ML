```python
model_1 = smf.ols(formula="MPG~HP+VOL+WT+SP",data=df).fit()

print("AIC = ",model_1.aic)
print("BIC = ",model_1.bic)
print("R2 = ",model_1.rsquared)
print("AdjR2 = ",model_1.rsquared_adj)
```

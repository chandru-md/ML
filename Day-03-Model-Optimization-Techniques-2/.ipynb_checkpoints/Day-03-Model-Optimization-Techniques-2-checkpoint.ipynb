```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
data = pd.read_csv("cars.csv")
```

```python
data
```

```python
cars = data.copy()
```

```python
cars
```

```python
cars["log_HP"] = np.log(cars["HP"])
cars['log_VOL'] = np.log(cars["VOL"])
cars["log_SP"] = np.log(cars["SP"])
cars["log_WT"] = np.log(cars["WT"])
```

```python
cars
```

```python
model_1 = smf.ols(formula="MPG~HP+VOL+SP+WT",data=cars).fit()
print("AIC",model_1.aic)
print("BIC",model_1.bic)
print("Rsquared",model_1.rsquared)
print("Rsquared_Adj",model_1.rsquared_adj)
```

```python
model_2 = smf.ols(formula="MPG~HP",data=cars).fit()
print("AIC",model_2.aic)
print("BIC",model_2.bic)
print("Rsquared",model_2.rsquared)
print("Rsquared_Adj",model_2.rsquared_adj)
```

```python
model_3 = smf.ols(formula="MPG~VOL",data=cars).fit()
print("AIC",model_3.aic)
print("BIC",model_3.bic)
print("Rsquared",model_3.rsquared)
print("Rsquared_Adj",model_3.rsquared_adj)
```

```python
model_4 = smf.ols(formula="MPG~SP",data=cars).fit()
print("AIC",model_4.aic)
print("BIC",model_4.bic)
print("Rsquared",model_4.rsquared)
print("Rsquared_Adj",model_4.rsquared_adj)
```

```python
model_5 = smf.ols(formula="MPG~WT",data=cars).fit()
print("AIC",model_5.aic)
print("BIC",model_5.bic)
print("Rsquared",model_5.rsquared)
print("Rsquared_Adj",model_5.rsquared_adj)
```
